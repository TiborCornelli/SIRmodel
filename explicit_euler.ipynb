```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
beta = 0.164
gamma = 1/14
N = 8*10**7

def f_1(S,I):
    return -beta * (S/N)*I

def f_2(S,I):
    return beta * (S/N) * I - gamma * I

def f_3(I):
    return gamma * I
```

```python
R_0 = 0
I_0 = 100
S_0 = N - I_0 - R_0
print(S_0)

def explicit_euler(t_0,T,n):
    t = np.linspace(t_0,T,n+1)
    dt = (T-t_0)/(n+1)

    S = np.zeros(n+1)
    I = np.zeros(n+1)
    R = np.zeros(n+1)

    S[0] = S_0
    I[0] = I_0
    R[0] = R_0
    print(S_0)
    for t_i in range(n):
        S[t_i+1] = S[t_i] +  dt * f_1(S[t_i], I[t_i])
        I[t_i+1] = I[t_i] +  dt * f_2(S[t_i],I[t_i])
        R[t_i+1] = R[t_i] +  dt * f_3(I[t_i])
    
    fig, ax = plt.subplots(figsize = (8,5))
    ax.margins(0.05)
    ax.set_title("Euler's Method, " + f"$t_0 = ${str(t_0)}, $T = ${str(T)}, $n = ${str(n)}", fontsize = 14)
    ax.plot(t,S, label="$S(t)$", color="green",)
    ax.plot(t,I, label="$I(t)$", color="red",)
    ax.plot(t,R, label="$R(t)$", color="blue",)
    plt.rc('legend',fontsize=16)
    ax.legend()
    plt.show()
```

```python
explicit_euler(0,365,365*5)
```
